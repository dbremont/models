# Multiplicative Effects

> **Multiplicative effects** refer to the phenomenon where the overall impact on a process or outcome is the result of the compounded product of multiple factors or variables, often leading to exponential growth or decay.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate the multiplicative effect using a dice model
np.random.seed(42)  # For reproducibility
```

```python
# Function to simulate the roll of a dice
def roll_dice():
    return np.random.randint(1, 7)

# Number of rolls
num_rolls = 10000
```

```python
# Simulate the multiplicative effect
results = []
for _ in range(num_rolls):
    product = 1
    for _ in range(10):  # Multiply the results of 10 dice rolls
        product *= roll_dice()
    results.append(product)
```

```python
# Plot the distribution
plt.hist(results, bins=50, density=True, log=True, color='skyblue', edgecolor='black')
plt.xlabel('Product Value')
plt.ylabel('Density (log scale)')
plt.title('Log-normal Distribution from Multiplicative Dice Rolls')
plt.show()
```
